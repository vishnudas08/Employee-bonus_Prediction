```
df= pd.read_csv('bonus_dataset.csv')
df.head()
```

```
performance= torch.tensor(df['performance'] ,dtype=torch.float32)
years_of_experience= torch.tensor(df['years_of_experience'] ,dtype=torch.float32)
projects_completed= torch.tensor(df['projects_completed'] ,dtype=torch.float32)

bonus= torch.tensor(df['bonus'] ,dtype=torch.float32)
```

```
w1= torch.rand(1, requires_grad=True)
w2 = torch.rand(1, requires_grad=True)
w3 = torch.rand(1, requires_grad=True)
b = torch.rand(1, requires_grad= True)
```

```
w1,w2,w3,b.item()
```

```
epochs=6000
learning_rate= 0.006
```

```
for epoch in range(epochs):
  predicted_bonus = w1 * performance + w2 * years_of_experience + w3 * projects_completed + b
  loss = ((predicted_bonus - bonus)**2).mean()
  loss.backward()
  with torch.no_grad():
    w1-= learning_rate * w1.grad
    w2-= learning_rate * w2.grad
    w3-= learning_rate * w3.grad
    b-= learning_rate * b.grad
  w1.grad.zero_()
  w2.grad.zero_()
  w3.grad.zero_()
  b.grad.zero_()
  if epoch % 100== 0:
    print(f"epoch: {epoch} ,loss: {loss.item():0.2f}")
```
